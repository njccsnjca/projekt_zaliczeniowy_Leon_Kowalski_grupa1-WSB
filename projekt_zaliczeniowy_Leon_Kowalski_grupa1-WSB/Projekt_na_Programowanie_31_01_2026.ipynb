```python
stopwords_url = "https://raw.githubusercontent.com/bieli/stopwords/master/polish.stopwords.txt"

```

```python
df["text_without_stopwords"] = ""
```

```python
for i in range(len(df)):
    text = str(df.loc[i, "text"]).lower()
    words = text.split()

    clean_words = []

    for word in words:
        if word not in stopwords_url:
            clean_words.append(word)

    df.loc[i, "text_without_stopwords"] = " ".join(clean_words)
```

```python
import spacy

nlp = spacy.load("pl_core_news_sm")
```

```python
persons_list = []
```

```python
persons_list = []

for text in df["text"].fillna(""):
    doc = nlp(text)
    osoby = []

    for ent in doc.ents:
        if ent.label_ == "persName":
            osoby.append(ent.text)
    persons_list.append(osoby)

```

```python
df["persons"] = persons_list
```

```python
df["persons"]
```

```python
places_list = []

for text in df["text"].fillna(""):
    doc = nlp(text)
    miejsca = []

    for ent in doc.ents:
        if ent.label_ == "placeName":
            miejsca.append(ent.text)
    places_list.append(miejsca)
```

```python
df["places"] = places_list
```
